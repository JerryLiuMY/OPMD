```python
from numpy import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
thirds = []
seconds = []
for mean in range(100): 
    a = random.normal(mean, mean/14, 10000)
    third = np.mean(a**3) - np.mean(a**2)*np.mean(a)
    second = np.mean(a)*(np.mean(a**2) - np.mean(a)*np.mean(a))
    thirds.append(third)
    seconds.append(second)
plt.plot(thirds, seconds)
```

```python
thirds = []
seconds = []
for mean in range(100): 
    a = random.poisson(mean, 10000)
    third = np.mean(a**3) - np.mean(a**2)*np.mean(a)
    second = np.mean(a)*(np.mean(a**2) - np.mean(a)*np.mean(a))
    thirds.append(third)
    seconds.append(second)
plt.plot(thirds, seconds)
```
